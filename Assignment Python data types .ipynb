## PROBLEM 2

A dictonary contains the marks of the students and there names, where names are keys and marks are values bymistake the teachere has given the lowest marks to the student who got the highest marks and gave highest mark to the student who actually got lowest mark. You have to correct the mistake and update the dictonary. 

```python
students_marks_dict = {'Shubham': 45, 'Krity': 98, 'Rachna':41, 'Aniket': 56, 'Vaibhav': 78, 'Rushikesh': 79}
```

```python
#Find out lowest mark
lowest_marks = min(students_marks_dict.values())
print(f"Lowest Marks : {lowest_marks}")

#Find out highest mark
highest_marks = max(students_marks_dict.values())
print(f"Highest Marks : {highest_marks}")

#Find which student got lowest mark
lowest_marks_name = min(students_marks_dict, key=students_marks_dict.get)
print("Student with Lowest Marks : ",lowest_marks_name)
#Find which student got highest mark
highest_marks_name = max(students_marks_dict, key=students_marks_dict.get)
print("Student with Highest Marks : ",highest_marks_name)
```

```python
#update the marks by giving highest mark to lowest mark student and lowest mark to highest mark student

temp=students_marks_dict[highest_marks_name]
students_marks_dict[highest_marks_name] = students_marks_dict[lowest_marks_name]
students_marks_dict[lowest_marks_name] = temp

print(temp)
print(students_marks_dict[highest_marks_name])
print(students_marks_dict[lowest_marks_name])

print(lowest_marks_name)
print(highest_marks_name)


```

```python
#print the updated dictonary
print(students_marks_dict)
```

## PROBLEM 4

Measures of Averages Mean of a group is defined as the sum of the elements in the group divided by the number of elements in the group. Median is the middle number in a sorted, ascending or descending, list of numbers. 

```python
# List of product prices
List_of_prices = [200, 340, 350, 230, 234, 540, 540, 340, 200, 340, 231, 342, 200, 234, 200, 200]

print(sum(List_of_prices))
```

FIND MEAN

```python
# find the mean of the product prices and print it

mean_price = sum(List_of_prices)/len(List_of_prices)
print("Mean Price is : ",mean_price)
```

FIND MEDIAN

```python
# find the median price of the product
sorted_list_of_prices=sorted(List_of_prices)
print(sorted_list_of_prices)
mid=int(len(sorted_list_of_prices)/2)
median_price = (sorted_list_of_prices[mid-1]+sorted_list_of_prices[mid])/2
print("Median Price is : ",median_price)
```

FIND MODE

```python
#Find the mode price in the list of prices
import statistics

mode_price = statistics.mode(List_of_prices)
print(mode_price)
```
