# <u>**Problem 1**</u>

### Plot the total profit across the months

* X label name = Month Number
* Y label name = Total profit



```
# Your code here
df= pd.read_csv('/content/drive/MyDrive/AllDataSets/company_sales_data.csv')
df.head()
```

```
x_label=df['month_number']
y_label=df['total_profit']

plt.plot(x_label,y_label,'ro-',linewidth=2,markersize=8)
plt.title('Monthly Profit of Company',{'fontsize':12,'fontweight':30})
plt.ylabel('Total Profit',{'fontsize':10,'fontweight':15})
plt.xlabel('Month Number',{'fontsize':10,'fontweight':15})
plt.rcParams['figure.figsize']=(8,8)
```

# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
#Create figure Object
ax= df.groupby(['month_number'])[['facewash','facecream']].mean().plot.bar(
    width=0.8,
    linewidth=1,
    edgecolor='black',
    figsize=(12,6),
    fontsize=10,
)

ax.set_title("Face Cream and Faewash Product Sales Month Wise",fontsize=20)

#Set x and y-labels
ax.set_xlabel("Month Number", fontsize = 15)
ax.set_ylabel("Product Sales", fontsize = 15)
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here

all_products = df.loc[:,['facecream','facewash','toothpaste','bathingsoap','shampoo','month_number','moisturizer']]

#SETTING THE MONTH NUMBERS AS THE INDEX
all_products = all_products.set_index('month_number')

#PLOTTING ALL THE PRODUCTS
plt.plot(all_products['facecream'],'ro-',linewidth=2,markersize=5)
plt.plot(all_products['facewash'],'bo-',linewidth=2,markersize=4)
plt.plot(all_products['toothpaste'],'go-',linewidth=2,markersize=4)
plt.plot(all_products['bathingsoap'],'yo-',linewidth=2,markersize=4)
plt.plot(all_products['shampoo'],color='black',marker='o',linewidth=2,markersize=4)
plt.plot(all_products['moisturizer'],'o-',linewidth=2,markersize=4)

#TITLE
plt.title("Products Sales Month Wise", fontsize = 20)

#ALL THE LEGENDS
plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','month_number','moisturizer'])

#set x and y-labels
plt.xlabel("Month Number", fontsize= 15)
plt.ylabel("Product Sales", fontsize= 15)

#FIGURE SIZE
plt.rcParams['figure.figsize']=(10,8)
```
